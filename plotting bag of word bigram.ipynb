```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
t = pd.read_csv('bigramwithcount1.csv')
```

```python
t.head()
```

```python
t.drop([ 'Time','Classification Report'], axis=1, inplace=True)
```

```python
t.head()
```

```python
t.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
t.head()
```

```python
t.drop(t.index[0])
```

```python
t
```

```python
tt=t.sort_values(by='Accuracy', ascending=True)
```

```python
tt
```

```python
plt.bar("Classifier","Accuracy", data =tt, color = "blue")
plt.xlabel("Classifier")
plt.xticks(rotation = 90)
plt.ylabel("Accuracy")
plt.title("Accuracy of bag of words bigram")
plt.show()
```
