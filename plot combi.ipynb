```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a = pd.read_csv('bigramwithcount1.csv')
```

```python
b = pd.read_csv('bigramwithtfidf1.csv')
```

```python
c = pd.read_csv('trigramwithcountvectoriser1.csv')
```

```python
a.drop([ 'Time','Classification Report'], axis=1, inplace=True)
```

```python
b.drop([ 'Time','Classification report'], axis=1, inplace=True)
```

```python
c.drop([ 'Time','Classification report'], axis=1, inplace=True)
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(16,10))
plt.plot("Classifier","Accuracy", data =a,color="Orange",linestyle="--"  , label="bag of word Bigram",linewidth=2)
plt.plot("Classifier","Accuracy", data =b,  label="Tf-idf bigram",linewidth=2)
plt.plot("Classifier","Accuracy", data =c,  label="bag of word trigram",linewidth=2,linestyle=":")
plt.xlabel("Classifier")
plt.xticks(rotation = 90)
plt.ylabel("Accuracy")
plt.title("Accuracy of Different feature Extraction ")
plt.legend()
plt.show()
```
